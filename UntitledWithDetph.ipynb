```python
def Triangulate(image, p):
    p = np.array(p)
    triangulation = Delaunay(p)
    img = image.copy()
    for tri in triangulation.simplices:
        dr = p[tri].reshape(-1, 2)
        dr = np.array(dr, dtype=np.int64)
        rr, cc = polygon(dr[:,0], dr[:,1])
        img[rr, cc] = np.sum(img[rr, cc], axis=0)/ len(img[rr, cc])
    return img



Triangulation = Delaunay(p)
plt.imshow(Triangulate(image[:,:,0:3], p))
plt.show()
io.imsave("./plusdepth.jpg", Triangulate(image[:,:,0:3], p))
```

```python
def Error(image, p):
    p = np.array(p)
    triangulation = Delaunay(p)
    img = image.copy()
    for tri in triangulation.simplices:
        dr = p[tri].reshape(-1, 2)
        dr = np.array(dr, dtype=np.int64)
        rr, cc = polygon(dr[:,0], dr[:,1])
        img[rr, cc] = np.sum(img[rr, cc], axis=0)/ len(img[rr, cc])
    return np.sum((img-image)**2)

def find_neighbors(pindex, triang):
    return triang.vertex_neighbor_vertices[1][triang.vertex_neighbor_vertices[0][pindex]:triang.vertex_neighbor_vertices[0][pindex+1]]

def PartError(image, p, i, base_triangulation):
    p = np.array(p)
    v = []
    v += list(p[list(find_neighbors(i, base_triangulation))])
    triangulation = Delaunay(v)
    img = np.copy(image)
    for tri in triangulation.simplices:
        dr = p[tri].reshape(-1, 2)
        dr = np.array(dr, dtype=np.int64)
        rr, cc = polygon(dr[:,0], dr[:,1])
        img[rr, cc] = np.sum(img[rr, cc], axis=0)/ len(img[rr, cc])
    return np.sum((img-image)**2)
```

```python
def LocateRemovable(image, p, triangulation):
    pool = Pool()
    p = list(p)
    i = np.random.randint(2, len(p) - 4)
    
    q1 = p[:]
    q = p[:]
    q2 = p[:]
    
    res1 = pool.apply_async(PartError, [image, q1, i - 1, triangulation])
    res = pool.apply_async(PartError, [image, q, i, triangulation])
    res2 = pool.apply_async(PartError, [image, q2, i + 1, triangulation])
    err = res.get()
    err1 = res1.get()
    err2 = res2.get()
    
    p = np.array(p)
    pool.close()
    return i - 1 + np.argmin([err1, err, err2])
```

```python
def Thinning(image, p, triangulation):
    i = LocateRemovable(image, p, triangulation)
    p = list(p)
    p.pop(i)
    p = np.array(p)
    triangulation = Delaunay(p)
    return (image, p, triangulation)
```

```python
def MainAlgo(image, contour_image, depth_image, NUMSEGMENTS, K, lst):
    image = np.append(image, depth_image.reshape(depth_image.shape[0], depth_image.shape[1], 1), axis=2)
    segments = slic(image, n_segments = NUMSEGMENTS, sigma = 3)
    p = []
    for i in range(0, segments.shape[0], 3):
        for j in range(0, segments.shape[1], 3):
            if(len(Counter(segments[i:i+3,j:j+3].reshape(-1).tolist())) >= 3):
                p.append([i + 1, j + 1])
    p += [[0,0], [0, segments.shape[1] - 1], [segments.shape[0] - 1, 0], [segments.shape[0] - 1, segments.shape[1] - 1]]
    p = np.array(p)
    Triangulation = Delaunay(p)
    l = len(p)
    for i in tqdm.tqdm(range(l - K)):
        image, p, Triangulation = Thinning(image, p, Triangulation)
        if (len(p) in lst):
            ret, thresh = cv2.threshold(contour_image, 127, 255, 0)
            im2, contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
            p = list(p)
            q = p[:]
            for contour in contours:
                approx = cv2.approxPolyDP(contour,1,True)
                approx = approx.reshape(-1, 2)
                c = np.array(approx[:, 0])
                approx[:,0] = approx[:, 1]
                approx[:,1] = c
                q += list(approx.reshape(-1, 2))
            q = np.array(q)
            p = np.array(p)
            yield Triangulate(image, q)
```

```python
for l in range(1001, 1101):
    image = img_as_float(io.imread("./Picture_Contours_Edges/Contours&depth/Source/" + str(l) + ".jpg"))
    contour_image = cv2.imread("./Picture_Contours_Edges/Contours&depth/Contour/" + str(l) + "-matte.png", cv2.IMREAD_GRAYSCALE)
    depth_image = img_as_float(io.imread("./Picture_Contours_Edges/Contours&depth/Depth/" + str(l) + "-depth.png"))
    gen = MainAlgo(image, contour_image, depth_image, 1500, 100, [600, 500, 400, 300, 250, 200, 150, 120, 100])
    for K in [600, 500, 400, 300, 250, 200, 150, 120, 100]:
        transformedImage = gen.__next__()
        io.imsave("./results/" + str(l) + "-" + str(K) + ".jpg", transformedImage)
```
